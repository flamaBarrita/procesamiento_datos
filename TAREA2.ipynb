Ejercicio 1

Escribe un programa que solicite un "número de la suerte". El programa debe imprimir un mensaje si el número ingresado no es un número entero (integer).

```python
valor = input()

print("Valor válido" if valor.isdigit() else "Valor inválido")
```

Ejercicio 2

Reescribe el ejercicio anterior para que, adicionalmente, imprima cuántos dígitos tiene el número (siempre que el número sea un entero).

```python
valor = input()

if valor.isdigit():
    print("Valor válido")
    print(len(str(valor)))
else:
    print("Valor no válido")
```

Ejercicio 3

Escribe un programa que solicite dos veces un número entero.

El programa debe imprimir el mayor de los dos números. Si los números son iguales, el programa debe indicarlo.

```python
valor1 = int(input())
valor2 = int(input())

if valor1 > valor2:
    print(f"El mayor es {valor1}")
elif valor2 == valor1:
    print ("Los números son iguales")
else:
    print(f"El mayor es el {valor2}")
```

Ejercicio 4

Escribe un programa que solicite dos veces un número entero.

El programa debe mostrar la suma de todos los enteros dentro del rango de esos dos números, incluyéndolos.

    Ejemplo: Si el usuario ingresa 10 y 15, la suma sería 75.

        $10 + 11 + 12 + 13 + 14 + 15 = 75$



```python
valor1 = int(input())
valor2 = int(input())

suma = 0
for i in range(valor1,valor2+1):
    suma += i

print(f"La suma es: {suma}")
```

Ejercicio 5

Pide al usuario que ingrese varios números en una sola línea de entrada.

Divide (split) los números en una lista.

Escribe un bucle (loop) que examine cada elemento de la lista y muestre solo aquellos que sean mayores a cero.

```python
numeros = input().split()

for i in numeros:
    num = int(i)
    if num > 0 :
        print(i)
```
