Ejercicio 3

Escribe un programa que acepte una cadena de texto del usuario.

Determina e imprime la siguiente información sobre la cadena:

    ¿Termina con un punto?

    ¿Contiene únicamente caracteres alfabéticos?

    ¿Hay una letra 'x' en la cadena?

    Crea e imprime una nueva cadena donde todas las letras e se cambien por E.


```python
cadena = input()
print("¿Termina con un punto? -> ", cadena.endswith('.'))
print("¿Contiene únicamente caracteres alfabéticos? -> ", cadena.isalpha())
print("¿Hay una letra 'x' en la cadena? -> ", 'x' in cadena)
print(cadena.replace("e","E"))
```

Ejercicio 5

Escribe una aplicación que solicite ingresar el radio de un círculo.

Acepta la entrada del usuario en una variable.

Calcula e imprime el área del círculo cuyo radio fue ingresado.

La fórmula para el área de un círculo es πr² (pi por el cuadrado del radio).

Usa 3.14159 como valor de pi.

```python
radio = input()
pi = 3.14159
calculo = pi * float(radio) **2
print(f"Área: {calculo}")
```
